```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline 

src = cv.imread('img/field.jpg')

# white range
lower_white = np.array([0,0,100])
upper_white = np.array([160, 100, 255])

hsv = cv.cvtColor(src,cv.COLOR_BGR2HSV)
mask = cv.inRange(hsv, lower_white, upper_white)
plt.imshow(mask)

dst = cv.Canny(src, 50, 200, None, 3)
lines = cv.HoughLines(mask, 1, np.pi / 180, 1000, None, 0, 0)


if lines is not None:
    for i in range(0, len(lines)):
        rho = lines[i][0][0]
        theta = lines[i][0][1]
        a = math.cos(theta)
        b = math.sin(theta)
        x0 = a * rho
        y0 = b * rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        cv.line(src, pt1, pt2, (0,255,0), 3, cv.LINE_AA)

# plt.imshow(dst)
plt.imshow(src)
```
